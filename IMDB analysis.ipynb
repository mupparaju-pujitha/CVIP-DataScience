## Classify movies according to their rating

```python
def rating(rating):
    if rating>=7.5:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
```

```python
DATA['rating_category']=DATA['Rating'].apply(rating)
```

### Display unique values of Genre

```python
DATA['Genre']
```

```python
genre=[]
for value in DATA['Genre']:
    genre.append(value.split(','))
genre
```

```python
g=[]
for i in genre:
    for j in i:
        g.append(j)
```

```python
g
```

```python
unique=[]
for i in g:
    if i not in unique:
        unique.append(i)
```

```python
print(unique)
print("\nNo.of unique genres:",len(unique))
```
